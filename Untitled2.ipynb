```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import torch

torch.set_num_threads(32)
torch.set_num_interop_threads(32)

plt.rcParams['pdf.use14corefonts'] = True
plt.rc('text', usetex=True)
```

```python
def softmax(x):
    y = x - np.max(x)
    y = np.exp(y)
    f_x = y / np.sum(y)
    return f_x
```

```python
data = {'error': [], 'a': [], 'mode': []}

for a in range(1, 102, 10):
    for i in range(100):
        Q1 = np.random.randn(a)
        Q2 = np.random.randn(a)
        print(Q1.shape, Q2.shape)
        
        Q = Q1[np.argmax(Q1)]
        data['error'].append(Q)
        data['a'].append(a)
        data['mode'].append('naive')
    
        Q = Q2[np.argmax(Q1)]
        data['error'].append(Q)
        data['a'].append(a)
        data['mode'].append('double')

        Q = np.mean(Q1 * softmax(Q2 / 0.1))
        data['error'].append(Q)
        data['a'].append(a)
        data['mode'].append('softmax')
        
        Q = np.mean(Q1 * softmax(Q2 * 10))
        data['error'].append(Q)
        data['a'].append(a)
        data['mode'].append('double-softmax')
        
fig = plt.figure(dpi=100)
ax = plt.subplot(111)
df = pd.DataFrame(data, columns=['error', 'a', 'mode'])
sns.barplot(ax = ax, x="a", y="error", hue="mode", data=df)
```
